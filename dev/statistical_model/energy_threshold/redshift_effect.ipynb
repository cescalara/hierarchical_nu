```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def power_law(E, Emin, alpha):
    return (alpha-1)/Emin * (E/Emin)**(-alpha)
```

```python
Emin = 1e3
Eth = 1e4
alpha = 2.2
z = 5.0
fac = (Eth/Emin)**(1-alpha)
E = np.logspace(np.log(Emin), np.log(1e3*Emin), base=np.e)

fig, ax = plt.subplots()
ax.plot(E, power_law(E, Emin, alpha) + power_law(E, (1+z)*Emin, alpha), 
        color='k', alpha=0.7, lw=3)
ax.plot(E, (power_law(E, Eth, alpha) + power_law(E, (1+z)*Eth, alpha)), 
        color='m', linestyle=':', lw=3)

ax.set_xscale('log')
ax.set_yscale('log')
```
