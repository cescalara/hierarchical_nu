```python
from pareto import truncated_pareto
```

```python
import numpy as np
alpha = 2.
xmin = 5.
xmax = 10.
```

```python
p = truncated_pareto(alpha, xmin, xmax)
```

```python
x = np.linspace(4,11,10000)
y = p.pdf(x)

from matplotlib import pyplot as plt
plt.plot(x,y,"k-")
plt.show()
```

```python
x = np.linspace(4,11,10000)
y = p.cdf(x)
from matplotlib import pyplot as plt
plt.plot(x,y,"k-")
plt.show()
```

```python
x = np.linspace(0,1,1000)
y = p.inv_cdf(x)
from matplotlib import pyplot as plt
plt.plot(x,y,"k-")
plt.show()
```

```python
nsamples = int(1.e7)
rvs = p.samples(nsamples)
plt.figure()
#plt.hist(np.log10(rvs), 50, normed=True, histtype='step')
plt.hist(rvs, 50, normed=True, histtype='step')
x = np.linspace(5,10,1000)
y = p.pdf(x)
plt.plot(x,y,'k--')
plt.show()
```
